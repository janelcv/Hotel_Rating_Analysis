# Extract Data 

```python
# Dependencies
import pandas as pd
import numpy as np
import sqlite3
import datetime
```

```python
# Create connections to database
conn = sqlite3.connect("Data/Hotels.db")

#Load the database table into a pandas dataframe
ratings = pd.read_sql_query("select * from ratings;", conn)
conn.close()

# Preview the dataframe
ratings.head()
```

```python
#Convert the reviews_date to year - Create a new column with year of date field "reviews_date"
ratings['year'] = pd.DatetimeIndex(ratings['reviews_date']).year
ratings.head()
```

```python
#Combine reviews_text & reviews_title to reviews
```

```python
ratings['reviews'] = ratings.reviews_text.str.cat(ratings.reviews_title)
ratings.head()
```

```python
# Selecting two columns year & reviews 
ratings.loc[:,'year':'reviews']
```

```python
#Group by the year and reviews two columns 
Groupby_year_reviews  = ratings.groupby(['year','reviews']).mean()
Groupby_year_reviews
Groupby_year_reviews.loc[:,'year':'reviews']
```
