# Step 1: ETL
The hotel ratings for USA properties were obtained from https://data.world. This jupyter notebook is used to clean the data in preparation for data exploration using machine learning approaches.

```python
# Dependencies
import pandas as pd
```

## Extract data

```python
# Load the csv file in a dataframe
path = "Data/Datafiniti_Hotel_Reviews.csv"

df = pd.read_csv(path)
df.head()
```

## Transform data

### Data cleaning

```python
# List down the column names
df.columns
```

```python
# Remove columns that will not be used in the analysis
df1 = df.drop(columns=["id","dateAdded","dateUpdated","keys","reviews.dateSeen",
                       "reviews.username","sourceURLs", "country", "postalCode"],axis=1)

# Preview the dataframe
df1.head()
```

```python
# Show the number of variables and the number of hotel reviews
df1.shape
```

### Separating metadata from ratings data

```python
# Select columns containing metadata
df2 = df1[["name", "categories", "primaryCategories", "address", "city", 
           "province", "latitude", "longitude", "websites"]]
metadata = df2.drop_duplicates()
metadata.head()
```

### Separating ratings data from metadata

```python
# Select columns containing ratings data
ratings = df1[["name", "reviews.date", "reviews.rating", "reviews.sourceURLs", "reviews.text", 
           "reviews.title", "reviews.userCity", "reviews.userProvince"]]
ratings.head()
```

```python
# Clean the review source URLs by retaining the website's host (www) and domain name (*.com)
ratings["reviews.sourceURLs"] = ratings["reviews.sourceURLs"].str.split("/")[:-1].str[2]
ratings.head()
```

```python
ratings = ratings.rename(index=str, columns={"reviews.date": "reviews_date", "reviews.rating": "reviews_rating","reviews.sourceURLs": "reviews_sourceURLs", "reviews.text": "reviews_text",
                                  "reviews.title": "reviews_title", "reviews.userCity": "reviews_userCity","reviews.userProvince": "reviews_userProvince",})
```

## Load data to SQLite database

```python
# Dependencies
import sqlite3
```

```python
# Create a database and connect to it
conn = sqlite3.connect("Data/Hotels.db")

# Create a database table each for ratings and for metadata
metadata.to_sql("metadata", conn, if_exists = "replace")
ratings.to_sql("ratings", conn, if_exists = "replace")
```

```python
# Preview metadata
pd.read_sql_query("select * from metadata limit 3;", conn)
```

```python
# Preview ratings
pd.read_sql_query("select * from ratings limit 3;", conn)
```

```python
# Close the connection
conn.close()
```
